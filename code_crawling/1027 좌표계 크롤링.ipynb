```python
def convert_address_to_coordinates(address):
    """
    입력받은 주소를 WGS84 좌표계 좌표로 변환
    """
    KAKAO_REST_API_KEY = "d89c853ebcefdd731422cbd0cb4cb9a4"
    url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + address
    header = {'Authorization': 'KakaoAK ' + KAKAO_REST_API_KEY}
 
    r = requests.get(url, headers=header)
 
    if r.status_code == 200:
        lng = float(r.json()["documents"][0]["address"]['x'])
        lat = float(r.json()["documents"][0]["address"]['y'])
    else:
        return None
    
    return lat, lng

# Test
convert_address_to_coordinates(zero["도로명 주소"].iloc[0]) 
```

```python
zero["y"] = ""
zero["x"] = ""

for i in range(46):
    yx = convert_address_to_coordinates(zero["도로명 주소"].iloc[i]) 
    zero["y"][i] = yx[0]
    zero["x"][i] = yx[1]
```

```python
zero["동"] = ""
 
for i in range(46):
    x_val = zero.x[i]
    y_val = zero.y[i]
    base_url = 'https://dapi.kakao.com/v2/local/geo/coord2regioncode.json?x='+str(x_val)+'&y='+str(y_val)+'&input_coord=WGS84'
    
    your_api_key = 'd89c853ebcefdd731422cbd0cb4cb9a4'
    headers = {'Authorization':'KakaoAK '+your_api_key}

    api_req = requests.get(base_url,headers=headers)
    zero["동"][i] = json.loads(api_req.text)["documents"][1]['region_3depth_name']
```

## heatmap 연습

```python
from folium.plugins import HeatMap
```

```python
data = pd.read_csv('서울특별시_금연구역_정보.csv',encoding="UTF-8")

#sido, lat, long변수 선택
data = data[['sgg_nm','lat','long']]

# 서초구 추출
data_sc = data[data['sgg_nm']=='서초구']


#lat, long 리스트 작성
lat = list(data_sc['lat'])
long = list(data_sc['long'])


center = [37.473552,127.030250]
seocho_map = folium.Map(location=center,zoom_strat=15)

heatmap = HeatMap(zip(lat,long),
                 min_opacity=0.1,
                 radius=20,
                 blur=25,
                 max_zoon = 4,
                 color='Red')

seocho_map.add_child(heatmap)



```
